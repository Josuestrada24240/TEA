```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#Comparación de diametro con la altura según parcelas
df = pd.DataFrame({'Parcela':['1', '2', '3', '4', '5'],
                   'DAP':[16.08, 15.76, 11.49, 16.39, 18.16],
                   'HT':[5.39, 5.44, 7.79, 7.61, 8.422]})

df = df.set_index('Parcela')
fig, ax = plt.subplots()
df.plot(ax = ax)
plt.show()

plt.rcParams['figure.figsize'] = [3,3]
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')

#Comparación de diametro con altura segun parcelas
fig, ax = plt.subplots(2, 2, sharey = True)
parcela = ['1', '2', '3', '4', '5']
parámetro = {'DAP':[16.08, 15.76, 11.49, 16.39, 18.16], 'HT':[5.39, 5.44, 7.79, 7.61, 8.422]}
ax[0, 0].plot(parcela, parámetro['DAP'], color = 'tab:orange')
ax[0, 1].plot(parcela, parámetro['HT'], color = 'tab:green')
ax[1, 0].bar(parcela, parámetro['DAP'], color = 'tab:orange')
ax[1, 1].bar(parcela, parámetro['HT'], color = 'tab:green')

plt.show()

plt.rcParams['figure.figsize'] = [4,3]
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# Porcentaje de árboles con caracteristicas deseables y no deseables 
x = [481,41]
colors = plt.get_cmap('GnBu')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
#Evaluación de abaundancia de especies
##Encino(228), Roble(179)


import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.linspace(-179, 228, 100)
y = np.linspace(-179, 228, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z, cmap='Greens')

plt.contourf(x, y, z, cmap='Greens')
plt.colorbar();

plt.rcParams['figure.figsize'] = [4,3]
plt.show()
```
